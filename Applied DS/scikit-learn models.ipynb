```python
from pycaret.datasets import get_data
from pycaret.classification import ClassificationExperiment

# Load a sample dataset
data = get_data('iris')

# Initialize the experiment
exp = ClassificationExperiment()

# Setup the experiment with the sample data
exp.setup(data, target='species')
```

```python
# List all models and their libraries
models = exp.models()
models
```

```python
sklearn_models = list(models[models['Reference'].str.contains('sklearn')].index)
```

```python
sklearn_models
```

```python
exclude_models = list(models[~models['Reference'].str.contains('sklearn')].index)
exclude_models
```
