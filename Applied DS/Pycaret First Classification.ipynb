````python
import numpy as np
import pandas as pd
from pycaret.classification import *
````

````python
data = pd.read_csv('https://raw.githubusercontent.com/psdbia/Class-Share/refs/heads/main/data/loan_defaults.csv')
data.head()
````

````python
data['default'].value_counts()
````

````python
s = setup(data, target = 'default', session_id = 42, fix_imbalance=True)   #fix_imbalance=True
````

````python
s.y_transformed.value_counts()
````

````python
s.compare_models(sort='MCC')
````

````python
lr_mdl = s.create_model('lr')
````

````python
lr_mdl
````

````python
#tuned_lr_mdl = s.tune_model(lr_mdl)
````

````python
evaluate_model(lr_mdl)
````

# Classification Plots

### Confusion Matrix
Visualizes the performance of a classification model by showing the actual vs. predicted classifications.
```python
plot_model(model, plot='confusion_matrix')
```

### AUC (Area Under the Curve)
Plots the ROC curve and calculates the area under it, which helps in evaluating the model's ability to distinguish between classes.
```python
plot_model(model, plot='auc')
```

### Precision-Recall Curve
Shows the trade-off between precision and recall for different threshold values.
```python
plot_model(model, plot='pr')
```

### Feature Importance
Displays the importance of each feature in the model.
```python
plot_model(model, plot='feature')
```

### Learning Curve
Plots the training and validation scores over different training set sizes to show how the model learns over time.
```python
plot_model(model, plot='learning')
```

### Validation Curve
Shows the model's performance over a range of hyperparameter values.
```python
plot_model(model, plot='vc')
```

### Residuals Plot
Visualizes the residuals (differences between actual and predicted values) to assess the model's fit.
```python
plot_model(model, plot='residuals')
```

### Manifold Learning
Uses techniques like t-SNE or UMAP to visualize high-dimensional data in a lower-dimensional space.
```python
plot_model(model, plot='manifold')
```

### Calibration Curve
Plots the predicted probabilities against the actual outcomes to assess the calibration of the model.
```python
plot_model(model, plot='calibration')
```

### Lift Curve
Shows the lift of the model, which is the ratio of the results obtained with and without the model.
```python
plot_model(model, plot='lift')
```

### Gain Chart
Visualizes the cumulative gain of the model.
```python
plot_model(model, plot='gain')
```

### KS Statistic Plot
Plots the Kolmogorov-Smirnov statistic to evaluate the separation between the positive and negative classes.
```python
plot_model(model, plot='ks')
```

### Cumulative Accuracy Profile (CAP)
Shows the cumulative accuracy of the model.
```python
plot_model(model, plot='cap')
```

### Decision Boundary
Visualizes the decision boundaries of the model.
```python
plot_model(model, plot='boundary')
```
